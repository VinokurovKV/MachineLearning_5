```python
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets

def plot_knn_bound(X, y, scaler=None, n_neighbors=10, xlim=(-15, 15), ylim=(-20, 20)):
    # step size in the mesh
    h = 0.05

    # Create color maps
    cmap_light = ListedColormap(['C0', 'orange', 'cyan', 'green'][:np.unique(y).shape[0]])
    cmap_bold = ['C0', 'orange', 'c', 'darkgreen'][:np.unique(y).shape[0]]

    x_min, x_max = xlim
    y_min, y_max = ylim
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    grid = np.c_[xx.ravel(), yy.ravel()]

    X_scaled = X # if scaler is None
    if scaler is not None:
        grid = scaler.transform(grid)
        X_scaled = scaler.transform(X)

    # we create an instance of Neighbours Classifier and fit the data.
    clf = neighbors.KNeighborsClassifier(n_neighbors, algorithm='brute')
    clf.fit(X_scaled, y)

    Z = clf.predict(grid)

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y,
                    palette=cmap_bold, alpha=1.0, edgecolor="black")
    plt.xlabel('x_1')
    plt.ylabel('x_2')
    plt.title('   {}-NN'.format(n_neighbors))
    plt.grid()
    plt.show()
```

   $1$-$NN$  $10$-$NN$

```python
plot_knn_bound(X, y, n_neighbors=1)
plot_knn_bound(X, y, n_neighbors=10)
```

```python
count_vec_scores = []
tf_idf_scores = []

for params, score in results.items():
    normalizer_name = params[0]
    if normalizer_name == "CountVectorizer":
        count_vec_scores.append(score)
    else:
        tf_idf_scores.append(score)

count_vec_mean = np.mean(count_vec_scores)
tf_idf_mean = np.mean(tf_idf_scores)

print("  :")
print(f"CountVectorizer:   = {count_vec_mean}")
print(f"TfidfVectorizer:   = {tf_idf_mean}")

if count_vec_mean > tf_idf_mean:
    print(f"\n  : CountVectorizer    {count_vec_mean}")
else:
    print(f"\n  : TfidfVectorizer    {tf_idf_mean:}")
```

```python
import matplotlib.pyplot as plt
from collections import defaultdict

neighbor_scores = defaultdict(list)

for params, score in results.items():
    n_neighbors = params[1]
    neighbor_scores[n_neighbors].append(score)

k_values = sorted(neighbor_scores.keys())
mean_scores = [np.mean(neighbor_scores[k]) for k in k_values]

plt.figure(figsize=(10, 6))
plt.plot(k_values, mean_scores, marker='o', linewidth=2, markersize=8)
plt.xlabel(' ')
plt.ylabel(' ')
plt.title('     ')
plt.grid(True, alpha=0.3)
plt.xticks(k_values)
plt.show()

print("   k:")
for k, score in zip(k_values, mean_scores):
    print(f"k = {k}:  = {score:.4f}")
```
